# Пример работы с классом RotaryTable

В данном документе показывается, как создать объект класса "Поворотный стол", а также пример построения графиков.

# Импорт библиотек

````python
import matplotlib.pyplot as plt # Импорт библиотеки для построения графиков
from RotaryTable import rotary_table_2d
````

# Создание объекта класс «Поворотный стол»

## Начальное положение

Кортеж содержащий два значения — начальное положение по первой и второй оси
```python
initial_position = (0, 100)
```

## Максимальная скорость поворота стола

Кортеж содержащий два значения — максимальная скорость поворота стола по первой и второй оси
```python
maximum_rate = (1000, 2000)
```

## Максимальное ускорение поворота стола

Кортеж содержащий два значения — максимальное ускорение поворота стола по первой и второй оси
```python
maximum_acceleration = (10, 10)
```

## Шаг времени

Переменная, которая задает шаг времени (оптимальное значение $10^{-3} — 10^{-4}$)
```python
DT = 0.0001
```

````python
initial_position = (0, 100)
maximum_rate = (1000, 2000)
maximum_acceleration = (10, 10)
DT = 0.0001
````

````python
my_table = rotary_table_2d(DT, initial_position, max_rate=maximum_rate, max_accel=maximum_acceleration)
````

# Нескольно тестовых движений

````python
my_table.POS(1, 150, 100)
my_table.PRS(1, 150, 200)
my_table.PRV(1, 200, 10, 1)
my_table.RAT(1, 15, 10, 7)
my_table.SIN(1, 15, 3, 0.5, 180)
````

````python
my_table.POS(2, 200, 250)
my_table.PRS(2, 150, 200)
my_table.PRV(2, 150, 500, 1)
my_table.RAT(2, 15, 10, 5)
my_table.SIN(2, 15, 3, 0.5, 180)
````

# Построение графиков

Если нужно экспортировать значения углов, скоростей и т.п., то можно использовать атрибуты с пометкой "*\_history" 

````python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5), dpi=200)

axes[0].plot(my_table.time_history,
             my_table.accel_history[0], label="Первая ось")
axes[0].plot(my_table.time_history,
             my_table.accel_history[1], label="Вторая ось")
axes[0].set_title("График углового ускорения", fontsize=15)
axes[0].set_xlabel("$t$", fontsize=15)
axes[0].set_ylabel(r"$\varepsilon(t)$", fontsize=15)

axes[1].plot(my_table.time_history,
             my_table.rate_history[0], label="Первая ось")
axes[1].plot(my_table.time_history,
             my_table.rate_history[1], label="Вторая ось")
axes[1].set_title("График угловой скорости", fontsize=15)
axes[1].set_xlabel("$t$", fontsize=15)
axes[1].set_ylabel(r"$\omega(t)$", fontsize=15)

axes[2].plot(my_table.time_history,
             my_table.pos_history[0], label="Первая ось")
axes[2].plot(my_table.time_history,
             my_table.pos_history[1], label="Вторая ось")
axes[2].set_title("График угловой координаты", fontsize=15)
axes[2].set_xlabel("$t$", fontsize=15)
axes[2].set_ylabel(r"$\varphi(t)$", fontsize=15)

for ax in axes:
    ax.grid()
    ax.legend(prop={'size': 10}, loc="upper left")

fig.suptitle("Графики параметров поворотного стола", fontsize=25)
fig.tight_layout()
````
